```python
fig=plt.figure()
fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.plot(unemploy[0:12]["DATE"],unemploy[0:12]["UNRATE"])
ax2.plot(unemploy[12:25]["DATE"],unemploy[12:25]["UNRATE"])
plt.show()
```

```python
fig=plt.figure()
fig=plt.figure(figsize=(30,15))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.plot(unemploy[0:12]["DATE"],unemploy[0:12]["UNRATE"])
ax2.plot(unemploy[12:25]["DATE"],unemploy[12:25]["UNRATE"])
plt.show()
```

```python
## Syntax for subplot - fig.add_subplot(m,n,p) - where m,n indicates the grid of the matrix of plots, & p defines the position of plot.
```

```python
fig=plt.figure()
fig=plt.figure(figsize=(30,15))
ax1=fig.add_subplot(4,4,7)
ax2=fig.add_subplot(4,4,8)
ax3=fig.add_subplot(4,4,1)
ax4=fig.add_subplot(4,4,2)
ax1.plot(unemploy[0:12]["DATE"],unemploy[0:12]["UNRATE"])
ax2.plot(unemploy[12:25]["DATE"],unemploy[12:25]["UNRATE"])
plt.show()
```
